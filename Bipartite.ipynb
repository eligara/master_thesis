```python
import os
import sys
import pandas as pd
import numpy as np
import networkx as nx
from matplotlib import pyplot as plt
```

```python
working_dir = "/Users/filippo/Developer/tesi"
os.chdir(working_dir)
dirs = os.listdir("data")
```

```python
df = pd.read_csv(("%s/mainTable.csv"%working_dir))
df.head()
```

```python
threshold = 0.
```

```python
G = nx.Graph()
```

```python
for g in df.loc[:,"gene"].values:
    G.add_node(g)
```

```python
for f in df.keys().values[1:]:
    G.add_node(f[:8])
```

```python
len(G.nodes)
```

```python
for g,_ in enumerate(df.loc[:,"gene"].values):
    for f in df.keys().values[1:]:
        if(df.loc[g,f]>threshold):
            G.add_edge(g,f)
```

```python
options = {'node_color': 'orange', "edge_color":'gray', "font_color": 'white', "font-family":"Helvetica", "font_size":'20', "font_style":"bold", 'node_size': 50, 'width': 0.8, 'with_labels': False}
```

```python
lay = nx.layout.spring_layout(G, k=0.2)
```

```python
fig = plt.figure()
nx.draw(G, pos=lay, **options)
plt.show()
```
