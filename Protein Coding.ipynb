```python
import mygene
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

# Download infos

```python
mg = mygene.MyGeneInfo()
```

```python
mg.getgenes(['ENSG00000242268'], 'name,symbol,refseq.rna,type_of_gene')
```

```python
df = pd.read_csv("genes.txt", header=None)
df.head()
```

```python
query = mg.getgenes([g[:15] for g in df[0].values], 'name,symbol,refseq.rna, type_of_gene')
```

```python
query[0]
```

```python
types = []
for t in query:
    try:
        types.append(t['type_of_gene'])
    except:
        types.append('unknown')
```

## count types

```python
utype, count = np.unique(np.array(types), return_counts=True)
```

```python
fig = plt.figure(figsize=(10,5))
plt.bar(utype, count, color='blue')
plt.ylabel("# genes", fontsize=16)
plt.show()
fig.savefig("genes_distr.pdf")
```

```python
np.sum(count[:-1])
```

```python
df_pc = pd.read_csv("genes_pc.txt", header=None)
```

```python
df_main = pd.read_csv("mainTable.csv")
genes = np.array([gene[:15] for gene in df_main.loc[:,'gene'].values])
df_main['gene'] = genes
df_main.head()
```

```python
df_main_pc = pd.DataFrame(columns=df_main.columns)
df_main_pc.to_csv("mainTable_pc.csv", index = False)
```
