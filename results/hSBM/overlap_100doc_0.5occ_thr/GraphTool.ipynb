```python
import graph_tool.all as gt
```

```python
g = gt.collection.data['karate']
g
```

```python
state = gt.minimize_nested_blockmodel_dl(g, overlap=True, B_min=50, nonoverlap_init=False)
```

```python
state.draw()
```

```python
level = state.levels[0]
```

```python
membership_size = []
for el in level.get_overlap_blocks()[0]:
    membership_size.append(len(el))
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure()
plt.hist(membership_size, range=(-0.5,5.5), bins=6, histtype='step')
plt.show()
```
